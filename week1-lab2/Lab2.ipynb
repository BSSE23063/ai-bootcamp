```python
# TODO: Word frequency counter using dictionary
text = "Python is simple yet powerful. Python is easy to learn."
# Split text and count frequency
word_freq = {}
words = text.split(" ")
for word in words:
    word = word.strip().lower() 
    if word:  
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
            
print(word_freq)            
```

```python
# TODO: Define a function that appends items to a list and observe mutability
def modify_list(my_list):
    my_list.append("new item")
    my_list.append("another item")
    my_list.pop()  # Remove the last item
    my_list[0] = "modified item"  # Modify the first item
    return my_list
    

lst = ["item1", "item2"]
modify_list(lst)
print(lst)
```
