```python
# **1. Series:**
#    - Create a Pandas Series with the following data: [10, 20, 30, 40, 50].
#    - Access the third element of the Series.
#    - What is the data type of the Series?


import pandas as pd
data=pd.Series([10, 20, 30, 40, 50])

print("The third element of the Series is:", data[2])
print("The data type of the Series is:", data.dtype)
```

```python
#    - Create a Pandas DataFrame with two columns: 'Name' and 'Age'.
#    - Add three rows to the DataFrame with the following data:
#      - Name: 'Alice', Age: 25
#      - Name: 'Bob', Age: 30
#      - Name: 'Charlie', Age: 35
#    - Access the 'Age' column of the DataFrame.
#    - What is the shape of the DataFrame?
dataframe=pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})

print("The 'Age' column of the DataFrame is:\n", dataframe['Age'])
print("The shape of the DataFrame is:", dataframe.shape)
```

```python
# **3. Indexing:**
#    - Using the DataFrame from question 2, access the row with the name 'Bob' using `loc`.
#    - Access the age of 'Alice' using `iloc`.
print("The row with the name of bob is:\n", dataframe.loc[dataframe['Name']=='Bob'])
print("The age of alice is:\n", dataframe.iloc[0]['Age'])
```

```python
# **4. File I/O:**
#    - Read the 'president_heights.csv' file into a DataFrame.
#    - What is the average height of the presidents?
#    - What is the name of the tallest president?
president_data= pd.read_csv('./data/president_heights.csv')
dataframe = pd.DataFrame(president_data)
# print("The dataframe is:\n", dataframe)
Average_height = dataframe['height(cm)'].mean()
print("The average height of the presidents is:", Average_height)

tallest_president = dataframe.loc[dataframe['height(cm)'].idxmax()]
print("The name of the tallest president is:", tallest_president['name'])
```

```python
# **5. Advanced:**
#    - Create a DataFrame with three columns: 'A', 'B', and 'C'.
#    - Fill the DataFrame with random numbers.
#    - Calculate the mean of each column.
#    - Add a new column 'D' which is the sum of columns 'A' and 'B'.
#    - Sort the DataFrame by column 'D' in descending order.
import numpy as np
df=pd.DataFrame({
    'A': np.random.rand(5),
    'B': np.random.rand(5),
    'C': np.random.rand(5)
})
print("The DataFrame with random numbers is:\n", df)

columns_mean = df.mean()
print("The mean of each column is:\n", columns_mean)

df['D'] = df['A'] + df['B']
print("The DataFrame with the new column is:\n", df)

sorted_df =df.sort_values(by='D',ascending=False)
print("The sorted DataFrame is:\n", sorted_df)
```
